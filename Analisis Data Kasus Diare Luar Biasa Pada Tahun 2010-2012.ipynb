# Data Kasus Diare Luar Biasa Pada tahun 2010-2012

```python
import pandas as pd
kasus_diare = pd.read_csv('processed-kasus-diare-luar-biasa-2010-2012.csv', delimiter=",", encoding="utf-8")
```

```python
kasus_diare
```

# Deskripsi Data
Data diatas berisi mengenai kasus diare luar biasa yang terjadi per provinsi. Data tersebut diambil pada tahun 2010-2012. Data bersumber dari Ditjen PP & PL, Kemenkes RI, tahun 2013. Berikut ini merupakan penjelasan mengenai Variabel yang digunakan pada Dataset ini:

a. kode_provinsi: Kode Provinsi

b. nama_provinsi: Nama Provinsi

c. tahun: Tahun

d. jumlah_penderita: Jumlah yang menderita diare luar biasa

e. jumlah_meninggal: Jumlah yang meninggal karena diare luar biasa

f. cfr: Case Fatality Rate (Persentase Jumlah Meninggal dengan Jumlah Penderita)

# Statistik Data

```python
# Jumlah penderita terbanyak pada kasus diare luar biasa dalam satu provinsi
a = kasus_diare.jumlah_penderita.max()
print ("Jumlah penderita terbanyak dalam satu provinsi= ", a, "orang")

# Jumlah kematian terbanyak pada kasus diare luar biasa dalam satu provinsi
b = kasus_diare.jumlah_meninggal.max()
print ("Jumlah meninggal terbanyak dalam satu provinsi= ", b, "orang")

# Presentase tertinggi Case Fatality Rate (CFR)
c = kasus_diare.cfr.max()
print ("Presentase tertinggi CFR= ", c, "%")
```

```python
# Jumlah kasus diare luar biasa seluruh provinsi per tahun 2010-2012
kasus_diare.groupby('tahun').sum()
```

```python
# Rata-rata jumlah terjadinya kasus diare luar biasa per provinsi selama tahun 2010-2012
kasus_diare.groupby('nama_provinsi').mean()
```

```python
# Rata-rata jumlah terjadinya kasus diare luar biasa per tahun 2010-2012 
kasus_diare.groupby('tahun').mean()
```

# Visualisasi Data

```python

ax = kasus_diare.groupby('tahun').mean()['jumlah_penderita']
ax = ax.plot(kind='bar', title ="Rata-Rata Jumlah Penderita Diare Berdasarkan Tahun", figsize=(8,8), legend=False, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = kasus_diare.groupby('nama_provinsi').mean()['jumlah_penderita']
ax = ax.plot(kind='bar', title ="Rata-Rata Jumlah Penderita Diare Setiap Provinsi pada Tahun 2010-2012", figsize=(12,5), legend=False, fontsize=12)
ax.set_xlabel("Nama Provinsi", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

# Analisis

   Dari hasil visualisasi nilai statistik pada data kasus diare luar biasa yang terjadi pada tahun 2010-2012, dapat dilihat bahwa jumlah kasus diare luar biasa pada tahun 2010 hingga 2012 mengalami penurunan yang cukup signifikan. Pada tahun 2010 terjadi kasus diare sebanyak 4204 kasus di Indonesia, kemudian pada tahun selanjutnya yaitu tahun 2011 terjadi 3003 kasus, dan pada tahun 2012 kasus diare luar biasa mengalami penurunan menjadi 1585 kasus. 
  
  Dari seluruh provinsi di Indonesia, daerah dengan jumlah kasus tertinggi terdapat pada provinsi Kepulauan Riau yang pernah mencapai 1426 kasus pada tahun 2011, kemudian dengan jumlah kasus diare tertinggi selanjutnya adalah provinsi Jawa Barat dan Jawa Timur. Namun terdapat juga beberapa provinsi yang tidak terserang kasus diare selama tahun 2010-2012 yaitu provinsi Bengkulu, Jakarta, Jambi, Kalimantan Barat, Kalimantan Selatan, Kalimantan Timur, Bangka Belitung, Maluku, Maluku Utara, dan NTB.
   
   Pada data ini dapat diambil kesimpulan bahwa kasus diare luar biasa yang kerap terjadi di beberapa provinsi di Indonesia cenderung mengalami penurunan tiap tahunnya. Hal ini menunjukkan bahwa tingkat kepedulian masyarakat terhadap kesehatan mulai meningkat, serta taraf kehidupan dan lingkungan yang bersih sudah menjadi perhatian masyarakat sehingga kasus penyakit seperti diare dapat dihindari.
